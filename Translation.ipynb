```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load dataset and dictionary

```python
df = pd.read_csv('data/extract/df.csv')
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
dicionario = pd.read_csv('data/extract/dicionario.csv')
```

```python
dicionario
```

```python
dicionario.rename(columns={'descrição':'desc'}, inplace=True)
```

```python
dicionario.codigo_da_variavel = dicionario.codigo_da_variavel.apply(lambda name: name.replace(' ', ''))
```

# Translate values

```python
df
```

```python
def translate(possibilities, value):
    if np.isnan(value):
        return value
    elif str(int(value)) in possibilities.keys():
        return possibilities[str(int(value))]
    else:
        return value


for col in df.columns.to_list(): 
    possibilities = dicionario[dicionario.codigo_da_variavel == col][[
        'tipo', 'categorias']].set_index('tipo').to_dict()['categorias']
    print(col)
    print (possibilities)
    df[col] = df[col].apply(lambda v: translate(possibilities, v))
```
