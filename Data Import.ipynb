# Data Cleaning

```python
columns = ['posicao inicial', 'tamanho','codigo_da_variavel', 'num', 'descrição', 'tipo', 'categorias']
translator = pd.DataFrame(columns=columns)

with open('data/PNS/2019/Microdados/Documentação/Dicionario_e_input_20210507/dicionario_PNS_microdados_2019.csv', 'r') as reader:
    
    for line in reader:
        row = line.replace('\n', '')
        row = row.split(';')
        
        print (row)
        if row[0] != '':
            save = row.copy()
            translator = translator.append({'posicao inicial':row[0],
                                            'tamanho':row[1],
                                            'codigo_da_variavel':row[2],
                                            'num':row[3],
                                            'descrição':row[4],
                                            'tipo':row[5],
                                            'categorias':row[6]},
                                           ignore_index=True)
        else:
            translator = translator.append({'posicao inicial':save[0],
                                            'tamanho':save[1],
                                            'codigo_da_variavel':save[2],
                                            'num':save[3],
                                            'descrição':save[4],
                                            'tipo':row[5],
                                            'categorias':row[6]},
                                           ignore_index=True)
```

```python
translator.to_csv('data/extract/dicionario.csv', index=False)
```
