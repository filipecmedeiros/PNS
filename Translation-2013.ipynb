# Translate values

```python
df
```

```python
def translate(possibilities, value):
    if len(possibilities) == 1:
        return value
    else:
        if np.isnan(value):
            return value
        elif str(int(value)) in possibilities.keys():
            return possibilities[str(int(value))]
        else:
            return value


for col in df.columns.to_list()[995:]: 
    possibilities = dicionario[dicionario.codigo_da_variavel == col][[
        'tipo', 'categorias']].set_index('tipo').to_dict()['categorias']
    print(col)
    print (possibilities)
    df[col] = df[col].apply(lambda v: translate(possibilities, v))
```
